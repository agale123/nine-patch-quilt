```python
import pandas as pd
from random import random
from math import sqrt
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
from matplotlib.patches import Polygon
```

```python
cols = 7
rows = 17
# Quilt dimensions in inches
w = 55
h = 74 * (rows + 1) / rows
```

```python
# Initialize the plot
plt.clf()
plt.rcParams["figure.figsize"] = (10, 10)
fig, ax = plt.subplots()
plt.xlim([1, 20])
plt.ylim([55, 74])

zig_zag_color_map = {1: "#003F5C", 2: "#472B62", 3: "#ccb31f", 4: "#1c6e73"}

# Returns the background color first and then the three colors for the triangle
def get_colors(r, c):
    if r== 14:
        return "#ccb31f" # Yellow
    elif (r < 14 and c % 2 == 0) or (r > 14 and c % 2 ==1):
        return "#1c6e73" # Teal
    else:
        return "#003F5C" # Blue


def draw_triangle(r, c):
    y = r * (h / rows)
    x = c * (w / cols)

    # The thickness of the strips without margins
    a = 2
    # The height of the triangle
    b = w / cols

    fg = get_colors(r, c)

    if (r + c) % 2 == 0:
        # Pointing left
        # Dark
        ax.add_patch(
            Polygon(
                [
                    (x - b / sqrt(3) + a / sqrt(3)+0.16, y + a / sqrt(3)), #Problematic
                    (x - 2 * b / 3 + 2 * a, y),
                    (x + b / 3, y + (b - 4) / sqrt(3)),
                    (x + b / 3, y + b / sqrt(3)),
                ],
                color=fg,
                alpha=1,
                ec=fg,
                lw=1,
            )
        )
        # Medium
        ax.add_patch(
            Polygon(
                [
                    (x + b / 3, y + (b - 4) / sqrt(3)),
                    (x + b / 3 - a, y + (b - 4 - a) / sqrt(3)),
                    (x + b / 3 - a, y - b / sqrt(3) + a / sqrt(3)),
                    (x + b / 3, y - b / sqrt(3)),
                ],
                color=fg,
                alpha=0.7,
                ec=fg,
                lw=1,
            )
        )
        # Light
        ax.add_patch(
            Polygon(
                [
                    (x + b / 3 - a, y - b / sqrt(3) + a / sqrt(3)),
                    (x + b / 3 - a, y - (b - 4 - a) / sqrt(3)),
                    (x - b / sqrt(3) + a / sqrt(3)+0.16, y + a / sqrt(3)), #Problematic
                    (x - 2 * b / 3 + 0.05, y),
                ],
                color=fg,
                alpha=0.4,
                ec=fg,
                lw=1,
            )
        )
    else:
        # Pointing right
        # Dark
        ax.add_patch(
            Polygon(
                [
                    (x + b / sqrt(3) - b / 3 - a / sqrt(3)-0.16, y + a / sqrt(3)),
                    (x + b / 3 - 2 * a, y),
                    (x - 2 * b / 3+0.05, y + (b - 4) / sqrt(3)), # Problematic
                    (x - 2 * b / 3+0.05, y + b / sqrt(3)),
                ],
                color=fg,
                alpha=1,
                ec=fg,
                lw=1,
            )
        )
        # Medium
        ax.add_patch(
            Polygon(
                [
                    (x - 2 * b / 3 + a, y - b / sqrt(3) + a / sqrt(3)),
                    (x - 2 * b / 3 + a, y - (b - 4 - a) / sqrt(3)),
                    (x + b / sqrt(3) - b / 3 - a / sqrt(3)-0.16, y + a / sqrt(3)),
                    (x + b / 3, y),
                ],
                color=fg,
                alpha=0.7,
                ec=fg,
                lw=1,
            )
        )
        # Light
        ax.add_patch(
            Polygon(
                [
                    (x - 2 * b / 3+0.05, y + (b - 4) / sqrt(3)), # Problematic
                    (x - 2 * b / 3 + a, y + (b - 4 - a) / sqrt(3)),
                    (x - 2 * b / 3 + a, y - b / sqrt(3) + a / sqrt(3)),
                    (x - 2 * b / 3+0.05, y - b / sqrt(3)),
                ],
                color=fg,
                alpha=0.4,
                ec=fg,
                lw=1,
            )
        )


for i in range(13, 16):
    for j in range(1, 3):
        draw_triangle(i, j)

# Hide ticks
plt.gca().set_aspect("equal")
plt.axis("off")
plt.savefig("logo.png", bbox_inches="tight", transparent=True, dpi=400)
```
