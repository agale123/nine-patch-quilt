```python
import pandas as pd
import random
from math import sqrt
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
from matplotlib.patches import Polygon
```

```python
COLS = 4
ROWS = 4

COLORS = [
    "#7A1B27",
    "#B65226",
    "#CA6B16",
    "#C5852F",
    "#CC9526",
    "#989629",
    "#3D6536",
    "#015852",
    "#366E7F",
    "#22424D",
    "#301A3F",
    "#572141",
]
```

```python
def plot_square(ax, x, y, dx, dy, colors):
    ax.add_patch(Rectangle((x, y), dx, dy, color="black", ec=None))

    for i in range(0, len(colors)):
        ax.add_patch(
            Polygon(
                [
                    (x + dx * (4 - i) / 8, y + dy * (1 - i / 8)),
                    (x + dx * (7 - i) / 8, y + dy * (1 - i / 8)),
                    (x + dx * (7 - i) / 8, y + dy * (7 - i) / 8),
                    (x + dx * (3 - i) / 8, y + dy * (7 - i) / 8),
                ],
                color=COLORS[colors[i]],
                ec=None,
            )
        )
        ax.add_patch(
            Polygon(
                [
                    (x + dx * (1 - i / 8), y + dy * (4 - i) / 8),
                    (x + dx * (1 - i / 8), y + dy * (7 - i) / 8),
                    (x + dx * (7 - i) / 8, y + dy * (7 - i) / 8),
                    (x + dx * (7 - i) / 8, y + dy * (3 - i) / 8),
                ],
                color=COLORS[colors[i]],
                ec=None,
            )
        )


def try_pattern(patterns):
    plt.clf()
    plt.rcParams["figure.figsize"] = (COLS * 2, ROWS * 2)
    fig, ax = plt.subplots()
    ax.set_xlim([0, COLS])
    ax.set_ylim([0, ROWS])

    for r in range(0, ROWS):
        for c in range(0, COLS):
            plot_square(
                ax,
                c + (1 if c > 1 else 0),
                r + (1 if r > 1 else 0),
                1 if c < 2 else -1,
                1 if r < 2 else -1,
                patterns[r * 4 + c],
            )
```

```python
try_pattern(
    [
        # Row 0
        [7, 5, 3, 0],  # 0, 0
        [1, 5, 8, 11],  # 0, 1
        [0, 2, 4, 6],  # 0, 2
        [8, 9, 10, 11],  # 0, 3
        # Row 1
        [8, 9, 10, 11],  # 1, 0
        [0, 2, 4, 6],  # 1, 1
        [1, 5, 8, 11],  # 1, 2
        [7, 5, 3, 0],  # 1, 3
        # Row 2
        [7, 5, 3, 0],  # 2, 0
        [1, 5, 8, 11],  # 2, 1
        [0, 2, 4, 6],  # 2, 2
        [8, 9, 10, 11],  # 2, 3
        # Row 3
        [8, 9, 10, 11],  # 3, 0
        [0, 2, 4, 6],  # 3, 1
        [1, 5, 8, 11],  # 3, 2
        [7, 5, 3, 0],  # 3, 3
    ]
)
```

```python
try_pattern(
    [
        # Row 0
        [8, 9, 10, 11],  # 0, 0
        [1, 4, 7, 10],  # 0, 1
        [0, 3, 6, 9],  # 0, 2
        [8, 9, 10, 11],  # 0, 3
        # Row 1
        [8, 9, 10, 11],  # 1, 0
        [0, 2, 4, 6],  # 1, 1
        [1, 3, 5, 7],  # 1, 2
        [7, 5, 3, 1],  # 1, 3
        # Row 2
        [7, 5, 3, 1],  # 2, 0
        [1, 3, 5, 7],  # 2, 1
        [0, 2, 4, 6],  # 2, 2
        [8, 9, 10, 11],  # 2, 3
        # Row 3
        [8, 9, 10, 11],  # 3, 0
        [0, 3, 6, 9],  # 3, 1
        [1, 4, 7, 10],  # 3, 2
        [8, 9, 10, 11],  # 3, 3
    ]
)
```

```python
def get_shuffle():
    vals = list(range(0,11))
    random.shuffle(vals)
    return sorted(vals[:4])
    
try_pattern([get_shuffle() for _ in range(16)])
```
