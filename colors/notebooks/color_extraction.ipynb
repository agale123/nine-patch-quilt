```python
import cv2
import numpy as np
import pandas as pd
```

```python
FABRIC_LINE = "art_gallery_pure_solids"
crop = 50
```

```python
# Read in data
df = pd.read_csv("../web/data/" + FABRIC_LINE + ".csv")
df
```

```python
# Extract the average rgb colors from an image
def extract_rgb(src):
    myimg = cv2.imread("../web/images/" + FABRIC_LINE + "/" + src)
    cropped = list(map(lambda x: x[crop:-crop], myimg[crop:-crop]))
    avg_color_per_row = np.average(cropped, axis=0)
    (b, g, r) = np.round(np.average(avg_color_per_row, axis=0)).astype(int)
    return (r, g, b)
```

```python
# Apply extraction to all images
df["rgb"] = df['src'].apply(extract_rgb)
df
```

```python
# Write data to csv
df.to_csv("../web/data/" + FABRIC_LINE + ".csv", index=False)
```
