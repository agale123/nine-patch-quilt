```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
```

```python
plt.rcParams["figure.figsize"] = (20, 20)
plt.figure(dpi=200)
fig, ax = plt.subplots()
plt.gca().set_aspect('equal')

ax.set_xlim([-2, 50])
ax.set_ylim([61, -2])


def add_block(origin, width, height, label):
    ax.add_patch(Rectangle(origin, width, height, facecolor="white", edgecolor="blue"))
    plt.text(origin[0] + 0.5 * width, origin[1] + 0.5 * height, label, ha="center")


# Australia
add_block((0, 0), 13.5, 13.25, "Kookaburra") # ok
add_block((0, 13.25 + 2), 13.5, 13.25, "Cockatoo") # ok

add_block((13.5 + 2, 0), 13.25, 20.25, "Koala") # ok
add_block((13.5 + 2, 20.25 + 2), 13.25, 6.25, "Platypus") # ok

add_block((13.5 + 13.25 + 2 + 2, 0), 6.5, 5.75, "Echidna") # ok
add_block((13.5 + 13.25 + 2 + 2, 5.75 + 2), 6.5, 5.5, "Lorikeet") # ok
add_block((13.5 + 13.25 + 6.5 + 2 + 2 + 2, 0), 8.5, 13.25, "Emu") # ok
add_block((13.5 + 13.25 + 2 + 2, 5.75 + 5.5 + 2 + 2), 17, 13.25, "Kangaroo") # ok

# New Zealand 30.5
add_block((0, 30.5), 17, 13.25, "Combo") # ok
add_block((0, 30.5 + 13 + 2), 8.5, 13.25, "Owl") # ok
add_block((8.5 + 2, 30.5 + 13.25 + 2), 6.5, 5.5, "Fantail") # ok
add_block((8.5 + 2, 30.5 + 13.25 + 2 + 5.5 + 2), 6.5, 5.75, "Tui") # ok

add_block((17 + 2, 30.5), 13.25, 6.25, "Kiwi") # ok
add_block((17 + 2, 30.5 + 6.25 + 2), 13.25, 20.25, "Tui") # ok

add_block((17 + 2 + 13.25 + 2, 30.5), 13.5, 13.25, "Pukeko") # ok
add_block((17 + 2 + 13.25 + 2, 30.5 + 13.25 + 2), 13.5, 13.25, "Kereru") # ok
```
